```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from typing import List
```

```python
df = pd.read_csv("data.csv")
df
```

```python
df.info()
```

```python
df_reduced = df.drop(columns=['Submitted Time'])
df_reduced
```

```python
question_ids = {i + 1: question for i, question in enumerate(df_reduced.columns[1:])}
question_ids
```

```python
df_renamed = df_reduced.rename(columns={v: k for k, v in question_ids.items()})
df_renamed
```

```python
quantitative_indices = ["S.NO"] + [i for i in range(1,11)]
quantitative_indices
```

```python
df_quantitative = df_renamed[quantitative_indices]
df_quantitative
```

```python
def convert_answer_to_integer(answer: str, answers: List[str]) -> int:
    answers_lookup = {answer: i for i, answer in enumerate(answers)}
    return answers_lookup[answer.strip()] + 1
```

## Quantitative (1-10)

```python
df_quantitative_mapped = df_quantitative.copy()
for i in range(1,11):
    df_quantitative_mapped[i] = df_quantitative[i].apply(lambda x: convert_answer_to_integer(x, ["stimmt nicht zu", "stimmt manchmal zu", "stimmt meistens zu", "stimmt zu"]))
df_quantitative_mapped
```

```python
fig, ax = plt.subplots(2, 5, figsize=(36,6))
fig.subplots_adjust(top=1.5)
for i in range(0,10):
    ax_x = i // 5
    ax_y = i % 5
    counts, bins, bars = ax[ax_x, ax_y].hist(df_quantitative_mapped[i+1].values, bins=4)
    ax[ax_x, ax_y].axvline(x = df_quantitative_mapped[i+1].mean(), color = "black", linestyle = "dotted")
    ax[ax_x, ax_y].set_title(question_ids[i+1])
    ax[ax_x, ax_y].title.set_size(9)
    bin_centers = 0.5 * np.diff(bins) + bins[:-1]
    ax[ax_x, ax_y].set_xticks(bin_centers, ["stimmt\nnicht zu", "stimmt\nmanchmal zu", "stimmt\nmeistens zu", "stimmt zu"])
    ax[ax_x, ax_y].tick_params(labelsize=9)
    for b, bar in enumerate(bars):
        bar.set_facecolor(["#004D40", "#00796B", "#009688", "#4DB6AC"][b])
fig.savefig("questions_1_to_10.png", bbox_inches='tight', dpi=300)
```

### Grand Mean

mean over all questions

```python
print(f"Mean: {np.mean(df_quantitative_mapped[range(1,11)].values)}")
```

## Quantitative (11)

```python
labels_11 = ["nicht wichtig", "neutral", "wichtig", "sehr wichtig"]
q_17_mapped = df_renamed[11].apply(lambda x: convert_answer_to_integer(x, labels_11))

print(" | ".join([f"{l} : {i}" for i, l in enumerate(labels_11)]))
print(f"Mean: {q_17_mapped.mean()}")
print(f"Median: {q_17_mapped.median()}")
print(f"SD: {q_17_mapped.std()}")
print(f"SE: {q_17_mapped.sem()}")
```

## Quantitative (17)

```python
labels_17 = ["Nein, ich sehe keinen Bedarf/ Vorteil daran", "es wäre mir egal", "ich habe keinen Bedarf, aber finde es sinnvoll", "Ja, sehr"]
q_17_mapped = df_renamed[17].apply(lambda x: convert_answer_to_integer(x, labels_17))

print(" | ".join([f"{l} : {i}" for i, l in enumerate(labels_17)]))
print(f"Mean: {q_17_mapped.mean()}")
print(f"Median: {q_17_mapped.median()}")
print(f"SD: {q_17_mapped.std()}")
print(f"SE: {q_17_mapped.sem()}")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(14,5))
labels_17_breaks = ["Nein, ich\nsehe keinen\nBedarf/ Vorteil\ndaran", "es wäre\nmir egal", "ich habe\nkeinen Bedarf,\naber finde es\nsinnvoll", "Ja, sehr"]
labels = [labels_11, labels_17_breaks]
for i, v in enumerate([11,17]):
    counts, bins, bars = ax[i].hist(df_quantitative_mapped[i+1].values, bins=4)
    ax[i].axvline(x = df_quantitative_mapped[i+1].mean(), color = "black", linestyle = "dotted")
    ax[i].set_title(question_ids[v])
    ax[i].title.set_size(9)
    bin_centers = 0.5 * np.diff(bins) + bins[:-1]
    ax[i].set_xticks(bin_centers, labels[i])
    ax[i].tick_params(labelsize=9)
    for b, bar in enumerate(bars):
        bar.set_facecolor(["#0D47A1", "#1976D2", "#2196F3", "#64B5F6"][b])
fig.savefig("questions_11_17.png", bbox_inches='tight', dpi=300)
```
